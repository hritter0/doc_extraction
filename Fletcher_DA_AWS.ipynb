```python
def text_chunker(list_of_strings):
    new_list = []
    for item in list_of_strings:
        new_strings = textwrap.wrap(item, 1500)
        for item2 in new_strings:
            new_list.append(item2)
    return new_list
```

```python
train_pghs = text_chunker(train)
```

```python
len(train_pghs)
```

```python
tfidf = TfidfVectorizer(stop_words="english", ngram_range=(1,3), min_df=10)
tfidf_vecs = tfidf.fit_transform(train_pghs)
```

```python
tfidf_corpus = matutils.Sparse2Corpus(tfidf_vecs.transpose())
```

```python
id2word = dict((v, k) for k, v in tfidf.vocabulary_.items())
```

```python
lsi = models.LsiModel(tfidf_corpus, id2word=id2word, num_topics=300)
```

```python
lsi_corpus = lsi[tfidf_corpus]
```
